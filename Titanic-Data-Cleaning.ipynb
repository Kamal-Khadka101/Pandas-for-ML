```python
#importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the csv file
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.describe()
```

```python
print("Initial Missing Values:", df.isnull().sum())
```

```python
#Duplicated values hataune tarika
#df.drop_duplicates(inplace=True)
print("Duplicated Values if present", df.duplicated().sum())
```

```python
#Filling missing age values with median age
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
#Fill the missing values of Cabin with placeholder 'Unknown'
df['Cabin'].fillna('Unknown', inplace=True)
```

```python
#Filling missing Embarked values with most frequent values (mode)
mostFrequentValues = df['Embarked'].mode()[0]
df['Embarked'].fillna(mostFrequentValues, inplace=True)
```

```python
#Verifying the missing values
print("Clean dataset after the operation:\n", df.isnull().sum())
```

```python
import seaborn as sns
```

```python
#Check for outliners in 'Age' Column
print("Age Statistics:")
print(df['Age'].describe())
```

```python
#Plot a box plot to visualize outliners in 'Age'
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Age'])
plt.title("Age Distribution with Outliners")
```

```python
#Check for outliners in 'fare' column
print("\nFare Statistics:")
print(df['Fare'].describe())
```

```python
#Box plot to visualize outliners in 'Fare'
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Fare'])
plt.title("Fare Distribution with Outliners")
```

```python
#Import required library
from sklearn.preprocessing import LabelEncoder

#Encode converts string data to numerical one

#Create a label encoder object
le = LabelEncoder()

#Encode 'Sex' column
df['Sex'] = le.fit_transform(df['Sex'])

#Encode 'Embarked' column
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
# Create a new column 'FamilySize' by adding 'SibSp' and 'Parch'
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.head()
```
