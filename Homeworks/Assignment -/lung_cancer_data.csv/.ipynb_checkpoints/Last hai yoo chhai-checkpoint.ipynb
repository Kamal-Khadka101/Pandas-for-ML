```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Load the dataset
lung_cancer_data = pd.read_csv('lung_cancer_data.csv')
```

```python
# Display basic information about the dataset
print(lung_cancer_data.info())
```

```python
# Display the first few rows of the dataset
lung_cancer_data.head()
```

```python
lung_cancer_data.drop(['Patient_ID'], axis=1, inplace=True)
```

```python
# Summary statistics for numerical columns
lung_cancer_data.describe()
```

```python
# Summary statistics for categorical columns
lung_cancer_data.describe(include=['O'])
```

```python
# Define categorical and numerical columns
categorical_columns = ['Gender', 'Smoking_History', 'Tumor_Location', 'Treatment', 'Ethnicity', 'Insurance_Type', 'Family_History',
                       'Comorbidity_Diabetes', 'Comorbidity_Hypertension', 'Comorbidity_Heart_Disease',
                       'Comorbidity_Chronic_Lung_Disease', 'Comorbidity_Kidney_Disease', 'Comorbidity_Autoimmune_Disease',
                       'Comorbidity_Other']

numerical_columns = ['Age', 'Tumor_Size_mm', 'Survival_Months', 'Performance_Status', 'Blood_Pressure_Systolic', 'Blood_Pressure_Diastolic', 'Blood_Pressure_Pulse',
                     'Hemoglobin_Level', 'White_Blood_Cell_Count', 'Platelet_Count', 'Albumin_Level', 'Alkaline_Phosphatase_Level', 'Alanine_Aminotransferase_Level',
                     'Aspartate_Aminotransferase_Level', 'Creatinine_Level', 'LDH_Level', 'Calcium_Level', 'Phosphorus_Level', 'Glucose_Level', 'Potassium_Level',
                     'Sodium_Level', 'Smoking_Pack_Years']
```

```python
# Distribution of target variable
sns.countplot(data=lung_cancer_data, x='Stage')
plt.title('Distribution of Lung Cancer Stages')
plt.show()

# Distribution of numerical features
lung_cancer_data[numerical_columns].hist(figsize=(16, 12))
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
corr_matrix = lung_cancer_data[numerical_columns].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Feature Engineering

```python
# Encode categorical variables (except the target variable 'Stage')
encoder = OneHotEncoder(sparse=False, drop='first')
X_categorical = encoder.fit_transform(lung_cancer_data[categorical_columns])

# Define the target variable
y = lung_cancer_data['Stage']
```

```python
# Scale numerical features
scaler = StandardScaler()
X_numerical = scaler.fit_transform(lung_cancer_data[numerical_columns])
```

```python
# Combine numerical and categorical features
X = np.hstack([X_numerical, X_categorical])
```

Dimension Reduction with PCA

```python
# Apply PCA
pca = PCA(n_components=0.95)  # Keep 95% of the variance
X_pca = pca.fit_transform(X)

# Number of components chosen
print(f"Number of components chosen: {pca.n_components_}")
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
# Initialize models
models = {
    'SVM': SVC(),
    'Random Forest': RandomForestClassifier(),
    'KNN': KNeighborsClassifier(),
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier()
}

# Train and evaluate models
for name, model in models.items():
    # Perform cross-validation
    scores = cross_val_score(model, X_train, y_train, cv=5)
    print(f"{name} Accuracy: {scores.mean():.2f} (+/- {scores.std() * 2:.2f})")

# Example of hyperparameter tuning for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                           param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
print("Best parameters for Random Forest:", best_params)
```
